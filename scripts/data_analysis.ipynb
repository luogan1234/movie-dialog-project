```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
movie_data = np.load('movie_data_v1.npy', allow_pickle=True)[()]
vocab = np.load('vocab.npy')
genre_map = np.load('genre_map.npy', allow_pickle=True)[()]
movie_pretrain = np.load('movie_pretrain.npz', allow_pickle=True)
```

```python
print(len(movie_data))
print(len(vocab))
print(len(genre_map))
print(list(genre_map.keys()))
ratings = list(movie_pretrain['ratings'])
rating_set = list(set(ratings))
rating_set.sort()
print(rating_set)

# plt.hist(ratings)
# plt.savefig('movie_rating_hist.pdf')

plt.hist([int(x) for x in ratings], bins=range(2,11))
plt.savefig('movie_rating_hist_int.pdf')
```

```python
genre_list = {}
for key, value in genre_map.items():
    genre_list[value] = key
genre_list = [genre_list[x] for x in range(24)]
genres = movie_pretrain['genres']
genres = [x for g in genres for x in g]
print(genres[:5])
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1, 1, 1)
plt.hist(genres, bins=range(25))
ax.set_xticks(list(range(25)))
ax.set_xticklabels(genre_list, rotation=30, fontsize=10)
plt.savefig('genre_hist.pdf')
```
